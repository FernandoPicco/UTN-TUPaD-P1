Ejercicio 4

Escribí un programa que permita almacenar y consultar números telefónicos.

• Permití al usuario cargar 5 contactos con su nombre como clave y número como valor.

• Luego, pedí un nombre y mostrale el número asociado, si existe.

Ejemplo:

contactos = {"Juan": "123456", "Ana": "987654"}

#Consultar: "Juan" -> muestra "123456"

```python
# Crear un diccionario vacío
contactos = {}

for i in range(5): # permite al usuario almacenar 5 contactos (nombre + nro. telefónico)
    nombre = input(f"Ingrese el nombre del contacto {i+1}: ").strip().lower()
    numero = input(f"Ingrese el número telefónico de {nombre}: ").strip()
    contactos[nombre] = numero
print("Contactos almacenados correctamente: ")
print(contactos)

# Luego permitirle al usuario consultar sobre la existencia de un nombre y poder mostrarle el nro. telefónico.
consulta_nombre = input("Ingrese el nombre que desea consultar: ").strip().lower()

if consulta_nombre in contactos:
    print(f"El número telefónico de {consulta_nombre} es: {contactos[consulta_nombre]}")
else:
    print(f"No existe un contacto con el nombre '{consulta_nombre}'")

```
