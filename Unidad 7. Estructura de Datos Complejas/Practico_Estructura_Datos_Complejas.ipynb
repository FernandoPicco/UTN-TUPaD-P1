Ejercicio 1

Dado el diccionario precios_frutas

precios_frutas = {'Banana': 1200, 'Ananá': 2500, 'Melón': 3000, 'Uva':
1450}

Añadir las siguientes frutas con sus respectivos precios:

● Naranja = 1200

● Manzana = 1500

● Pera = 2300
