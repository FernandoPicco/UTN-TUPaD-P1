Ejercicio 1

Crear archivo inicial con productos: Crear un archivo de texto llamado productos.txt con tres productos. Cada línea debe tener: nombre,precio,cantidad

```python
with open("productos.txt", "w") as archivo:
    archivo.write("Destornillador, 200.04, 5\n")
    archivo.write("Abrazadera, 55.5, 10\n")
    archivo.write("Martillo, 123.24, 2\n")

    print("Archivo productos.txt creado correctamente.")
```

Ejercicio 2

Leer y mostrar productos: Crear un programa que abra productos.txt, lea cada línea, la procese con .strip() y .split(","), y muestre los productos en el siguiente formato:

Producto: Lapicera | Precio: $120.5 | Cantidad: 30

```python

print("****Productos****")
# Leer productos
with open("productos.txt", "r") as archivo:
    for linea in archivo:
        linea = linea.strip()
        partes = linea.split(",")

        nombre = partes[0]
        precio = partes[1]
        cantidad = partes[2]
# Mostrar por pantalla productos
        
        print(f"Producto: {nombre} | Precio: ${precio} | Cantidad:{cantidad}")
```

Ejercicio 3

Agregar productos desde teclado: Modificar el programa para que luego de mostrar los productos, le pida al usuario que ingrese un nuevo producto (nombre, precio, cantidad) y lo agregue al archivo sin borrar el contenido existente.

```python
# Pedir al usuario que ingrese un nuevo producto
print("\nIngrese un nuevo producto")
nombre_1 = input("Nombre: ").strip()
precio_1 = input("Precio: ").strip()
cantidad_1 = input("cantidad: ").strip()

#Para no borrar el contenido existente lo hago en modo append "a"

with open("productos.txt", "a") as archivo:
    archivo.write(f"{nombre_1}, {precio_1}, {cantidad_1}\n")

# Mostrar los productos agregados 

print("\nProducto agregado correctamente.")
```
