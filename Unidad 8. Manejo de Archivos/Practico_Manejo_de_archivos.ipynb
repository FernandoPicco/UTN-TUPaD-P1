Ejercicio 1

Crear archivo inicial con productos: Crear un archivo de texto llamado productos.txt con tres productos. Cada línea debe tener: nombre,precio,cantidad

Ejercicio 2

Leer y mostrar productos: Crear un programa que abra productos.txt, lea cada línea, la procese con .strip() y .split(","), y muestre los productos en el siguiente formato:

Producto: Lapicera | Precio: $120.5 | Cantidad: 30

```python

print("****Productos****")
# Leer productos
with open("productos.txt", "r") as archivo:
    for linea in archivo:
        linea = linea.strip()
        partes = linea.split(",")

        nombre = partes[0]
        precio = partes[1]
        cantidad = partes[2]
# Mostrar por pantalla productos
        
        print(f"Producto: {nombre} | Precio: ${precio} | Cantidad:{cantidad}")
```

Ejercicio 3

Agregar productos desde teclado: Modificar el programa para que luego de mostrar los productos, le pida al usuario que ingrese un nuevo producto (nombre, precio, cantidad) y lo agregue al archivo sin borrar el contenido existente.

```python
# Pedir al usuario que ingrese un nuevo producto
print("\nIngrese un nuevo producto")
nombre_1 = input("Nombre: ").strip()
precio_1 = input("Precio: ").strip()
cantidad_1 = input("cantidad: ").strip()

#Para no borrar el contenido existente lo hago en modo append "a"

with open("productos.txt", "a") as archivo:
    archivo.write(f"{nombre_1},{precio_1},{cantidad_1}\n")

# Mostrar los productos agregados 

print("\nProducto agregado correctamente.")
```

Ejercicio 4

Cargar productos en una lista de diccionarios: Al leer el archivo, cargar los datos en una lista llamada productos, donde cada elemento sea un diccionario con claves:

nombre, precio, cantidad.

```python
# Nueva lista vacía llamada "productos"
productos = []

# Se abre el archivo para leer
with open("productos.txt", "r") as archivo:
    for linea in archivo:
        linea = linea.strip()
        nombre, precio, cantidad = linea.split(",")
    # se crea el diccionario de datos

        producto = {
            "nombre": nombre,
            "precio": float(precio),
            "cantidad": int(cantidad)
        }
    # Se agrega el diccionario a la lista

        productos.append(producto)

# Mostramos por pantalla el resultado
print("Lista de productos cargada correctamente.")
for producto in productos:
    print(producto)
```

Ejercicio 5

Buscar producto por nombre: Pedir al usuario que ingrese el nombre de un producto. Recorrer la lista de productos y, si lo encuentra, mostrar todos sus datos. Si no existe, mostrar un mensaje de error.

```python
# Solicitar al usuario que ingrese el nombre del producto

nombre_producto_buscado = input("Ingrese el producto buscado: ").strip()

#declaramos una bandera para saber si lo encontramos
encontrado = False

for producto in productos:
    if producto['nombre'].lower() == nombre_producto_buscado.lower():
        print("\nProducto encontrado:")
        print(f"Nombre: {producto['nombre']}")
        print(f"Precio: ${producto['precio']}")
        print(f"Cantidad: {producto['cantidad']}")
        encontrado = True
        break

if not encontrado:
    print("El producto ingresado no se encuentra en la lista.")

        
```

Ejercicio 6

Guardar los productos actualizados: Después de haber leído, buscado o agregado productos, sobrescribir el archivo productos.txt escribiendo nuevamente todos los productos actualizados desde la lista.

```python
# Nueva lista vacía llamada "productos"
productos = []

# Se abre el archivo para leer
with open("productos.txt", "r") as archivo:
    for linea in archivo:
        linea = linea.strip()
        nombre, precio, cantidad = linea.split(",")
    # se crea el diccionario de datos

        producto = {
            "nombre": nombre,
            "precio": float(precio),
            "cantidad": int(cantidad)
        }
    # Se agrega el diccionario a la lista

        productos.append(producto)

# Mostramos por pantalla el resultado
print("Lista de productos cargada correctamente.")
for producto in productos:
    print(producto)

    
# Sobrescribir el archivo "productos.txt"

with open("productos.txt", "w") as archivo:
    for producto in productos:
        linea = f"{producto['nombre']},{producto['precio']},{producto['cantidad']}"
        archivo.write(linea)
print("\nArchivo -productos.txt- actualizado correctamente.")
```
